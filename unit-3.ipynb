```python
import pandas as pd
import plotly.express as px
data={'marks':[40,41,42,44,42,43,41,48,49,45,46,48],
     'gender':['M','F','M','M','F','M','M','F','F','F','M','M']}
df=pd.DataFrame(data)
df
```

```python
pd.get_dummies(df,drop_first=True)
```

Feture creation

```python
import numpy as np
df['result']=np.where(df['marks']>43,'pass','fail')
df
```

```python
df['grade']=np.where(df['marks']>47,'A',
                     np.where(df['marks']>45,'B',
                     np.where(df['marks']>43,'C','D')))
df
```
