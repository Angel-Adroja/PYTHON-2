# data cleaning

```python
data=[[1,np.nan,3,4,np.nan],
     [2,np.nan,np.nan,4,5],
     [4,3,np.nan,np.nan,np.nan],
     [4,2,5,6,4],
     [9,np.nan,5,4,5]]
df=pd.DataFrame(data)
df
```

```python
print(df.mean())
print(df.median())
print(df.mode())
```

```python
df.isnull()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean())
```

```python
df.fillna(0)
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='all')
```

```python
df.dropna(subset=[2,3])
```

```python
df.dropna(thresh=3)
```

```python
df.dropna(thresh=3,inplace=True)
```

```python
df.reset_index(drop=True)
```

```python
df.drop(1)
```

```python
df.drop(2,axis=1)
```

# Drop Duplicate values


```python
data=[[1,1,2,1,4],
     [2,1,2,4,3],
     [1,1,2,1,4],
     [3,4,3,2,1],
     [1,1,2,2,3],
     [1,1,2,1,4]]
df=pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
df.duplicated(keep='last')
```

```python
df.duplicated(subset=[0,1,2])
```

```python
df.drop_duplicates()
```
