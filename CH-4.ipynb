```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test,y_pred)) #sum(/Ytest-Ypred)/n
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))#sum(/Ytest-Ypred)^2/n
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))#1-sum(Ytest-Ypred)/sum(Ytest-Ymean)
```

```python
df=pd.read_csv('Marksheet.csv')
df.head()
```

```python
df.info()
```

```python
df['TEST-3 MARKS']=df['TEST-3 MARKS'].replace('AB',0)
df['TEST-4 MARKS']=df['TEST-4 MARKS'].replace('AB',0)
```

```python
df['TEST-3 MARKS']=df['TEST-3 MARKS'].astype('float')
df['TEST-4 MARKS']=df['TEST-4 MARKS'].astype('float')
```

```python
df['Total Marks']=df['TEST-1 MARKS']+df['TEST-2 MARKS']+df['TEST-3 MARKS']+df['TEST-4 MARKS']
df.head()
```

```python
import numpy as np
df['Grade']=np.where(df['Total Marks']>85,'A',
                    np.where(df['Total Marks']>75,'B',
                            np.where(df['Total Marks']>65,'C',
                                    np.where(df['Total Marks']>50,'D',
                                            np.where(df['Total Marks']>35,'E','F')))))
df.head()
```

```python
df.groupby('Div')['Total Marks'].mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

x=df[['TEST-1 MARKS']]
y=df['Total Marks']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(model.coef_)
print(model.intercept_)
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
r2=[]
for i in range(0,51):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    model=LinearRegression()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    r2.append(r2_score(y_test,y_pred))
plt.plot(r2)
print(max(r2))
print(r2.index(max(r2)))
```

```python
#1. Import required libraries.
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#2. Load RealEstate dataset, create a dataframe and check datatypes of its attributes using appropriate method.
df = pd.read_csv("Real estate.csv")
print(df.dtypes)

#3. Remove ‘No’ column from the dataframe.
df.drop('No',inplace =True, axis=1)


#4. Check for any null values in features using appropriate method.
df.isnull().sum()

#5. Create feature variables x and y as given above. 
x = df.drop("Y house price of unit area",axis=1)
y = df['Y house price of unit area']

#6. Create training and testing sets of feature variables with 70% of data for training and with random state of 110.
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state = 110)

#7. Create and fit regression model using appropriate method.
lr = LinearRegression()
lr.fit(x_train, y_train)

#8. Use testing set created in step 6 to find and print the prediction of the outcome.
y_pred = lr.predict(x_test)
print(y_pred)

#9. Find and print coefficient and mean squared error of the regression model.
print(lr.coef_)
print(lr.intercept_)
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
df = pd.read_csv("Ice_cream selling data.csv")
print(df.dtypes)


x = df[['Temperature (°C)']]
y = df['Ice Cream Sales (units)']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)

lr = LinearRegression()
lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)
print(y_pred)

print(lr.coef_)
print(lr.intercept_)
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))

import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()

import seaborn as sns
sns.regplot(x= x, y=y)
```
